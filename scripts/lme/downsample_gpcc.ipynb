```python
import xarray as xr
import pathlib
import xesmf as xe

DATA_FP = pathlib.Path(r"/glade/work/kcarr/lme_data")
```

```python
## load GPCC data
pr_gpcc = xr.open_dataset(DATA_FP / "precip_mon_025_v2020.nc").compute()

## resample to match CESM2 data
time_coder = xr.coders.CFDatetimeCoder(use_cftime=True)
pr = xr.open_dataset(DATA_FP / "snr" / "pr_split_deseason.nc", decode_times=time_coder)
grid = pr[["lat", "lon"]]

## do regridding
regridder = xe.Regridder(pr_gpcc, grid, "bilinear")
pr_gpcc_lores = regridder(pr_gpcc)

## save to file
pr_gpcc_lores.to_netcdf(DATA_FP / "precip_gpcc_lores.nc")
```
